```python
import pandas as pd
import glob
import re
```

```python
files = glob.glob('Alignments/*.Meg')
files
```

```python
def parse_meg(file):
    d = dict()
    file = open(file)
    organism = str()
    seq = str()
    gene = ''
    i = 0
    for line in file:
        if '#' in line and 'mega' not in line:
            if len(seq) > 0 and len(organism) > 0:
                d[i] = [organism,gene,seq]
                i += 1
            line = line.replace('\n', '').replace('#', '')
            pattern = r'_[A-Za-z0-9\-]*(_alt_version_\d)?$'
            organism = re.split(pattern, line)[0]
            gene = line.replace(organism, '')
            if gene[0] == '_':
                gene = gene[1:len(gene)]
            seq = ''
        else:
            seq = seq + line.replace('\n', '') 
    return d
```

```python
alignment_dfs = []
for file in files:
    d = parse_meg(file)
    df = pd.DataFrame.from_dict(d, orient='index', columns = ['organism', 'gene', 'alignment'])
    alignment_dfs.append(df)
alignment_df = pd.concat(alignment_dfs)
alignment_df

```
